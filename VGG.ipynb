# VGG16

```python
import numpy as np
from keras.applications import vgg16
from keras.preprocessing import image
```

```python
model = vgg16.VGG16(weights='imagenet')
```

```python
img = image.load_img('images/spoon.jpeg', target_size=(224,224))
#we specify 2 2 4 bc vgg use this image size
img 
```

```python
#convert to numpy array
arr = image.img_to_array(img)
arr.shape
```

```python
#expand dimension
arr = np.expand_dims(arr, axis=0)
arr.shape
```

```python
#preprocessing
arr = vgg16.preprocess_input(arr)
arr
```

```python
#predict
preds = model.predict(arr)
"""each of these numbers corresponds to the probability that the image that has been recognized is one of the 1000 categories 
in ImageNet"""
preds
```

```python
#predictions for top 5 categories
vgg16.decode_predictions(preds, top=5)
```

## using another image

```python
img2 = image.load_img("images/fly.jpeg", target_size=(224,224))
img2
```

```python
arr2 = image.img_to_array(img2)
arr2 = np.expand_dims(arr2, axis=0)
arr2 = vgg16.preprocess_input(arr2)
preds2 = model.predict(arr2)
vgg16.decode_predictions(preds2, top=5)
```
